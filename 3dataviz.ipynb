```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
car_crashes = sns.load_dataset("car_crashes")
car_crashes.head()
```

```python
sns.scatterplot(x="total", y="speeding", data=car_crashes)
plt.xlabel('Total Crashes')
plt.ylabel('Speeding Crashes')
plt.title('Relationship between Total Crashes and Speeding Crashes')
```

```python
sns.kdeplot(car_crashes['alcohol'])
plt.xlabel('Alcohol-related Crashes')
plt.ylabel('Density')
plt.title('Density Plot of Alcohol-related Crashes')
```

```python
sns.displot(car_crashes['alcohol'])
plt.xlabel('Alcohol-related Crashes')
plt.ylabel('Density')
plt.title('Distribution of Alcohol-related Crashes')
```

```python
sns.ecdfplot(car_crashes['alcohol'])
plt.xlabel('Alcohol-related Crashes')
plt.ylabel('ECDF')
plt.title('ECDF Plot of Alcohol-related Crashes')
```

```python
sns.regplot(x="total", y="alcohol", data=car_crashes)
plt.xlabel('Total Crashes')
plt.ylabel('Alcohol-related Crashes')
plt.title('Regression Plot of Total Crashes and Alcohol-related Crashes')
```

```python
sns.pairplot(car_crashes)
plt.show()
```

```python
plt.subplots(figsize = (16, 10))
sns.barplot(x = 'total', y = 'abbrev', data = car_crashes, orient = 'h')
plt.xlabel('Total number of accidents')
plt.ylabel('States')
plt.show()
```
